```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
AC_thrust = pd.read_csv('直流電流計測.csv', sep=',')

plt.figure(figsize=(10, 5))
plt.plot(AC_thrust['current'], AC_thrust['thrust'], marker='o')

plt.xlabel('current')
plt.ylabel('thrust')

plt.grid(True)

plt.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=go.Scatter(
        x=AC_thrust['current'],
        y=AC_thrust['thrust'],
         error_y=dict(
            type='percent', # value of error bar given as percentage of y value
            value=50,
            visible=True)
    ))
fig.show()
```
